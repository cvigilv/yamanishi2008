```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
DS = "../DS"
NETWORKS = ["e", "ic", "gpcr", "nr"]
FINGERPRINTS = ["fp2", "fp3", "fp4", "maccs"]
```

```python
for n in NETWORKS:
    for f in FINGERPRINTS:
        DS_file = f"{DS}/{n}_DS.{f}.txt"
        A = np.loadtxt(DS_file)
        plt.imshow(A)
        plt.title(f"{n} - {f}")
        plt.show()
```
